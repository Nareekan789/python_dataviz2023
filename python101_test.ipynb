## HW1
คำนวณว่า ช่วงเวลาต่อไปนี้ ห่างกันที่ ชั่วโมง กี่นาที กี่วินาที(print ออกมาให้สวยงาม)
- 12:30:15-13:41:07
- 09:11:50-21:30:15
- 19:25:30-03:10:15

````python
time_all = '12:30:15', '13:41:07' , '09:11:50', '21:30:15' ,'19:25:30', '03:10:15'
````

````python
#ข้อที่ 1
# 12:30:15-13:41:07
from datetime import time, datetime
time_a = time_all[0]
time_b = time_all[1]

# เปลี่ยนเวลาจาก string เป็น datetime
t1 = datetime.strptime(time_a, "%H:%M:%S")
t2 = datetime.strptime(time_b, "%H:%M:%S")

# หาความต่างของเวลา
delta = t2 - t1
str_ab = str(delta)
sp_ab = str_ab.split(':')
print(f"ดังนั้น เวลา {time_all[0]}น. กับเวลา {time_all[1]}น. ห่างกันอยู่ {sp_ab[0]}  ชั่วโมง {sp_ab[1]} นาที {sp_ab[2]} วินาที")
````

````python
#ข้อที่ 1
# 12:30:15-13:41:07
q1 = '12:30:15 - 13:41:07'
q2 = q1.split("-")
Q1 = q2[0].split(":")
Q2 = q2[1].split(":")
#---- แปลงเลขเป็นวินาที -----
hours1 = int(Q1[0])*3600
minutes1 = int(Q1[1])*60
seconds1 = int(Q1[2])
hours2 = int(Q2[0])*3600
minutes2 = int(Q2[1])*60
seconds2 = int(Q2[2])
#---- -----
f1 =  hours1 + minutes1 + seconds1#เอาวินาทีรวมกัน
f2 =  hours2 + minutes2 + seconds2
f3 = (f2-f1)
rh = int(f3/3600)
rm = int((f3%3600)/60)
rs = int(f3%60)
print(f' จาก {q1} ซึ่งห่างกันได้ {rh}:{rm}:{rs}')
````

````python
#ข้อที่ 2
# 09:11:50-21:30:15
time_c = time_all[2]
time_d = time_all[3]

# เปลี่ยนเวลาจาก string เป็น datetime
t3 = datetime.strptime(time_c, "%H:%M:%S")
t4 = datetime.strptime(time_d, "%H:%M:%S")

# หาความต่างของเวลา
delta_cd = t4 - t3
str_cd = str(delta_cd)
sp_cd = str_cd.split(':')
print(f"ดังนั้น เวลา {time_all[2]}น. กับเวลา {time_all[3]}น. ห่างกันอยู่ {sp_cd[0]}  ชั่วโมง {sp_cd[1]} นาที {sp_cd[2]} วินาที")
````

````python
q1 = '09:11:50 - 21:30:15'
q2 = q1.split("-")
Q1 = q2[0].split(":")
Q2 = q2[1].split(":")
hours1 = int(Q1[0])*3600
minutes1 = int(Q1[1])*60
seconds1 = int(Q1[2])
hours2 = int(Q2[0])*3600
minutes2 = int(Q2[1])*60
seconds2 = int(Q2[2])
f1 =  hours1 + minutes1 + seconds1
f2 =  hours2 + minutes2 + seconds2
f3 = (f2-f1)
rh = int(f3/3600)
rm = int((f3%3600)/60)
rs = int(f3%60)
print(f' จาก {q1} ซึ่งห่างกันได้ {rh}:{rm}:{rs}')
````

````python
#ข้อที่ 3
# 19:25:30-03:10:15
time_e = time_all[4]
time_f = time_all[5]

# เปลี่ยนเวลาจาก string เป็น datetime
t5 = datetime.strptime(time_e, "%H:%M:%S")
t6 = datetime.strptime(time_f, "%H:%M:%S")

# หาความต่างของเวลา
delta_ef = t5 - t6
str_ef = str(delta_ef)
sp_ef = str_ef.split(':')
print(f"ดังนั้น เวลา {time_all[4]}น. กับเวลา {time_all[5]}น. ห่างกันอยู่ {sp_ef[0]}  ชั่วโมง {sp_ef[1]} นาที {sp_ef[2]} วินาที")
````

````python
#ข้อ1 วิธีที่ 2 : รายการที่แพงที่สุดและรายการที่ถูกที่สุด
min_data1 = data1['Approved_buget'].min()
max_data1 = data1['Approved_buget'].max()
data1[(data1['Approved_buget'] == max_data1)| (data1['Approved_buget'] == min_data1)]
````

````python
#ข้อ1 วิธีที่ 3 : รายการที่แพงที่สุดและรายการที่ถูกที่สุด
result1= data1.groupby('Item_name')['Approved_buget'].aggregate(['max'])
re1 = result1[(result1['max'] == 28500000) | (result1['max'] == 0)]
re1.rename(columns = {'max' : 'รายการที่แพงที่สุดและถูกที่สุด'})
````

````python
#ข้อ2 รายการขุดลอกคูคลองของ'สำนักงานเขตคลองสาน'และ 'สำนักงานเขตคลองสามวา'
data1[(data1['Department'] == 'สำนักงานเขตคลองสาน') | (data1['Department'] == 'สำนักงานเขตคลองสามวา')]
````

````python
##ข้อ2 : รายการขุดลอกคูคลองของ'สำนักงานเขตคลองสาน'และ 'สำนักงานเขตคลองสามวา'
df = pd.DataFrame(data1,columns=['Item_name','Department'])
df.set_index('Department', inplace=True)
result = df.loc[["สำนักงานเขตคลองสาน","สำนักงานเขตคลองสามวา"],['Item_name']]
result.rename(columns = {'Item_name' : 'รายการขุดลอกคูคลอง'})
````

````python
#ข้อ 3 วิธีที่1 จากข้อ2เขตไหนใช้งบประมาณมากกว่ากัน
KlongSan = data1[data1['Department'] =='สำนักงานเขตคลองสาน'][['Approved_buget']].sum()
KlongSam = data1[data1['Department'] =='สำนักงานเขตคลองสามวา'][['Approved_buget']].sum()

print(f'งบประมาณของสำนักงานเขตคลองสาน คือ {int(KlongSan):,} บาท')
print(f'งบประมาณของสำนักงานเขตคลองสามวา คือ {int(KlongSam):,} บาท')

print('')
result = KlongSan < KlongSam

if result is True:
  print('ดังนั้นงบประมาณสำนักงานเขตคลองสาน มากกว่า งบประมาณสำนักงานเขตคลองสามวา')
else :
  print('ดังนั้นงบประมาณสำนักงานเขตคลองสามวา มากกว่า งบประมาณสำนักงานเขตคลองสาน')
````
